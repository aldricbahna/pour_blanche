```python
def clean_column(col):
    col = col.strip().lower()
    col = col.replace("'", "")
    col = col.replace("-", " ")
    col = col.replace("(", "").replace(")", "")
    col = col.replace("/", " ")
    col = "_".join(col.split())
    return col
df=df0.copy()
df.columns = [clean_column(c) for c in df0.columns]

# --- 2. Replace N/A, NSP etc. with np.nan ---
df = df.replace(
    {
        "N/A": np.nan,
        "NA": np.nan,
        "n/a": np.nan,
        "na": np.nan,
        "NSP": np.nan,
        "nsp": np.nan,
        "Ne sait pas":np.nan,
        "": np.nan,
        "on":np.nan
    }
)
df
```

```python
from scipy.stats import chi2_contingency
stat, p_value, dof, expected = chi2_contingency(table)
print("\n--- Résultats du Test ---")
print(f"Statistique Chi-2 : {stat:.4f}")
print(f"P-value           : {p_value:.4e}") # Notation scientifique pour plus de précision
print(f"Degrés de liberté : {dof}")

print("\n--- Fréquences théoriques (attendues) ---")
# Utile pour vérifier les conditions d'application (doivent être > 5)
print(pd.DataFrame(expected, index=table.index, columns=table.columns))
```

```python
def convert_age(x):
    if x=='> 65 ans':
        return 70
    elif isinstance(x, str) and "-" in x:
        nums = "".join([c if c.isdigit() or c == "-" else " " for c in x]).split("-")
        try:
            a, b = map(int, nums)
            return (a + b) / 2
        except:
            return np.nan
    try:
        return float(x)
    except:
        return np.nan

df["tranche_age"] = df["tranche_age"].apply(convert_age)
df['tranche_age'].value_counts()
```
